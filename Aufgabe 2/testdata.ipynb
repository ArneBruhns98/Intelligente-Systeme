```python
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.patches as patches
%matplotlib inline
```

```python
df1 = pd.read_csv(r"testtemp.txt", sep=",")
df1 = df1[df1.recall < 1]
df1 = df1[df1.precision < 1]
df1 = df1[df1.fscore < 1]
```

```python
size = df1.size 
shape = df1.shape 

print("Size = {}\nShape ={}\n".format(size, shape))

df1[df1['fscore']==df1['fscore'].max()].sort_values(by=['fscore'], ascending=False)
```

```python
df1[df1['recall']==df1['recall'].max()].sort_values(by=['precision'], ascending=False)
```

```python
df1[df1['precision']==df1['precision'].max()].sort_values(by=['recall'], ascending=False)
```
