```python
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.patches as patches
%matplotlib inline
```

```python
df1 = pd.read_csv(r"data0.csv", header=None)
```

```python
size = df1.size 
shape = df1.shape 
print(df1.head(20))
print(df1.tail(20))

print("Size = {}\nShape ={}\n". 
format(size, shape)) 
```

```python
df2 = pd.read_csv(r"labels0.csv", header=None, names=['X', 'Y'])
```

```python
size = df2.size 
shape = df2.shape 
print(df2.head(20))
print(df2.tail(20))

print("Size = {}\nShape ={}\n". 
format(size, shape)) 
```

```python
plt.figure(figsize=(104, 32))
ax = plt.gca()
ax.set_aspect('equal')
figure = plt.gcf()
plt.imshow(df1,cmap='hot',interpolation='nearest')
plt.xlabel('X-Coord')
plt.ylabel('Y-Coord')
plt.colorbar()

# adding points from label
for index, row in df2.iterrows():
    rect = patches.Rectangle((row['Y'],row['X']),5,5,linewidth=1,edgecolor='black',facecolor='black')
    ax.add_patch(rect)

plt.savefig('heatmap.png')
plt.show()
```
